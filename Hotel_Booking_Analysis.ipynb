Exploratory Data Analysis(EDA):It refers to the critical process of performing initial investigations on data so as to discover patterns, to spot anomalies , to test hypothesis and to check assumptions with the help of summary statistics and graphical representations.The following are the various steps performed as a part of EDA:
*   Data preparation
*   Data cleaning
*   Univariate analysis
*   Bivariate analysis
*   Multivariate analysis

Data preparation: Firstly, we imported libraries and dataset, some of the libraries used are NumPy,Pandas,Seaborn,Matplotlib.Once the data is collected the process of data analysis begins.But the data has to be translated in appropriate form.This process is known as Data preparation.
Data cleaning:The raw data recieved in the dataset might not be directly suitable for analysis due to the presence of unwanted data like,duplicate values, null values, outliers etc.We need to handle them first before we proceed with further analysis.
Removing duplicates:The duplicate values in the dataset need to be removed so that we get accurate patterns and numbers regarding the analysis.
Handling Null/Missing values:We need to either reomve Null/Missing values or find alternatives to fill up the null values.
Univariate analysis:Univariate analysis focuses on analyzing each feature in the dataset independently.
Bivariate analysis:Bivariate analysis focuses on analyzing two features in the dataset.
Multivariate analysis:Multivariate analysis focuses on analyzing two or more features in the dataset.

After completing all the above processes, we have observed final conclusions from the patterns of the analysis and identified anomalies that can help increase the business objective.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Creating copy of the dataset
hotel_booking_df=hotel_booking.copy()
```

```python
#deleting all duplicate rows
hotel_booking_df.drop_duplicates(inplace=True)
```

```python
#deleting rows having null values in children column
hotel_booking_df.dropna(subset=['children'],inplace=True)
```

```python
#deleting rows having null values in country column
hotel_booking_df.dropna(subset=['country'],inplace=True)
```

```python
# Replacing all the null values in the agent column with 0
# 0 means agent ID was not available
hotel_booking_df['agent'].fillna(0,inplace=True)
```

```python
# droping company column as it contains 94% Null values
hotel_booking_df.drop(['company'],axis=1,inplace=True)
```

```python
# droping arrival_date_week_number column as it is redundant
hotel_booking_df.drop(['arrival_date_week_number'],axis=1,inplace=True)
```

```python
# Deleting row containing negative value of ADR(Average Daily Rate) i.e. -6.38
hotel_booking_df.drop(14969,inplace=True)
```

```python
# changing dtype of certain columns to category dtype
cat_lst=['hotel','meal','country','market_segment','distribution_channel','assigned_room_type','reserved_room_type','deposit_type','customer_type','reservation_status', 'is_canceled', 'is_repeated_guest']
for i in cat_lst:
  hotel_booking_df[i]=hotel_booking_df[i].astype('category')
```

```python
# changing dtype of certain columns to int dtype
int_lst=['children','agent']
for i in int_lst:
  hotel_booking_df[i]=hotel_booking_df[i].astype(int)
```

```python
# changing reservation_status_date column dtype from object to datetime
hotel_booking_df['reservation_status_date']= pd.to_datetime(hotel_booking_df['reservation_status_date'],format='%Y-%m-%d')
```

```python
# creating a new column named as 'Total_no._of_people'
hotel_booking_df['Total_no._of_people']=hotel_booking_df['adults']+hotel_booking_df['babies']+hotel_booking_df['children']
```

```python
# creating a new column named as 'total_no._of_nights'
hotel_booking_df['total_no._of_nights']=hotel_booking_df['stays_in_weekend_nights']+hotel_booking_df['stays_in_week_nights']
```

```python
#dropping the row containing value of ADR=5400
hotel_booking_df=hotel_booking_df[hotel_booking_df['adr']<5000]
```

```python
#checking the shape of the dataset after all manipulations
hotel_booking_df.shape
```

```python
hotel_booking_df.describe()
```

```python
hotel_booking_df.info()
```

### What all manipulations have you done and insights you found?

1.   Dropped all the duplicate rows.
2.   Deleted all rows having null values in children and country column because that was < 1%  of the data.So, it will not affect the analysis.
3.   Replaced all the null values in the agent column with 0. 0 means agent ID was not available.
4.   Dropped company column as it contains 94% Null values.
5.   Dropped arrival_date_week_number column as it was redundant.
6.   Deleted a row containing negative value of ADR(Average Daily Rate) i.e. -6.38
7.   Changed dtype of certain columns to category,int,datetime dtype.
8.   Created two new columns named as 'Total_no._of_people' and 'total_no._of_nights'.
9.   Dropped the row containing value of ADR=5400,which is a outlier.

By converting the 'reservation _status _date column to a different datatype, such as datetime, we can extract information based on quarters, months, or years. This allows us to analyze the booking patterns over different time periods.

Additionally, we introduced new columns to capture the total number of people involved in each booking. By aggregating the 'children, 'babies', and 'adults' columns into a new column called 'Total_no. _of_people', we can analyze the bookings based on the total number of individuals.

Another new column we added is 'total_no._of_nights', which represents the total number of nights for each booking. This allows us to examine the bookings from the perspective of the duration of stay.
